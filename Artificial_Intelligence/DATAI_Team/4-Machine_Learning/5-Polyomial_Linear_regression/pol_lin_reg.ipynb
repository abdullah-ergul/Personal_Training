```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("polynomial_regression.csv", sep=";")
```

```python
x = df.araba_max_hiz.values.reshape(-1,1)
y= df.araba_fiyat.values.reshape(-1,1)

plt.scatter(x,y)
plt.xlabel("Araba Max Hiz")
plt.ylabel("Araba Fiyat")
plt.show()
```

```python
lr = LinearRegression()
lr.fit(x,y)
```

```python
y_head = lr.predict(x)
plt.scatter(x,y)
plt.plot(x, y_head, color="red", label ="poly")
```

it's not good

```python
from sklearn.preprocessing import PolynomialFeatures

pol_reg = PolynomialFeatures(degree= 2)
x_polomial = pol_reg.fit_transform(x)
```

```python
lin_reg2 = LinearRegression()
lin_reg2.fit(x_polomial, y)

y_head2 = lin_reg2.predict(x_polomial)

plt.scatter(x,y)
plt.plot(x, y_head, color="red", label ="lin")
plt.plot(x, y_head2, color="green", label ="poly")

plt.legend
plt.show
```

```python
pol_reg2 = PolynomialFeatures(degree= 5)
x_polomial2 = pol_reg2.fit_transform(x)

lin_reg3 = LinearRegression()
lin_reg3.fit(x_polomial2, y)

y_head3 = lin_reg3.predict(x_polomial2)

plt.scatter(x,y)
plt.plot(x, y_head, color="red", label ="lin")
plt.plot(x, y_head2, color="green", label ="poly")
plt.plot(x, y_head3, color="black", label ="poly2")


plt.legend
plt.show()
```

better
