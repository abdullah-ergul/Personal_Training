```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv("linear_regression_dataset.csv", sep=";")
```

```python
plt.scatter(df.deneyim, df.maas)
plt.xlabel("Deneyim")
plt.ylabel("Maas")
```

```python
linear_reg = LinearRegression()
x = df.deneyim.values.reshape(-1,1)
y = df.maas.values.reshape(-1,1)

linear_reg.fit(x,y)
```

```python
#prediction
b0 = linear_reg.intercept_
print("b0: ", b0) # intercept, y
b1 = linear_reg.coef_
print("b1: ",b1)  # slope, eğim

print(linear_reg.predict([[11]])) # 1663 + 1138*11
```

```python
arr = np.array(range(16)).reshape(-1,1)
plt.scatter(x,y)
y_head = linear_reg.predict(arr)
plt.plot(arr, y_head, color="red")
```
